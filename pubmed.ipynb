```python
import datasets.pubmed as pubmed
import torch
from torch import optim

from GAT_model import GAT
from trainer import Trainer
import matplotlib.pyplot as plt
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)

if device.type == 'cuda':
    print(torch.cuda.get_device_name(0))
    print('Memory Usage:')
    print('Allocated:', round(torch.cuda.memory_allocated(0)/1024**3,1), 'GB')
    print('Cached:   ', round(torch.cuda.memory_reserved(0)/1024**3,1), 'GB')

pubmed_data = pubmed.load_pubmed('datasets/pubmed/', device=device)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)

if device.type == 'cuda':
    print(torch.cuda.get_device_name(0))
    print('Memory Usage:')
    print('Allocated:', round(torch.cuda.memory_allocated(0)/1024**3,3), 'GB')
    print('Cached:   ', round(torch.cuda.memory_reserved(0)/1024**3,3), 'GB')
```

```python
model = GAT(
    num_of_layers=2,
    num_heads_per_layer=[1, 1],
    num_features_per_layer=[pubmed.NUM_INPUT_FEATURES, 50, pubmed.NUM_CLASSES],
    add_skip_connection=False,
    bias=True,
    dropout=0.7,
)

optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-3)
trainer = Trainer(model=model, optimizer=optimizer, data=pubmed_data, return_logs=True)

logs_na = trainer.train(500)
```
